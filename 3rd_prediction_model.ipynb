```python
import numpy as np
import matplotlib.pyplot as mp
import pandas as pd
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
boston
```

```python
#describing the data
print(boston.DESCR)
```

```python
#transforming the datasets into a dataframe
data = pd.DataFrame(boston.data, columns = boston.feature_names)
data2 = pd.DataFrame(boston.target)
```

```python
data
```

```python
data2
```

```python
#spliting the data into 67% training and 33% testing
x_train, x_test, y_train, y_test = train_test_split(data, data2, test_size = 0.33, random_state = 42)
```

```python
reg = linear_model.LinearRegression()
```

```python
#training the model
reg.fit(x_train, y_train)
```

```python
x_test
```

```python
#printing the coefficients for each column
print("weights are ", reg.coef_)
# f(x) = mx + db + c = y
# where m is slope, c is intercept, x is independent varible, y is dependent variable
```

```python
#printing the intercept
print("intercept is ", reg.intercept_)
```

```python
#predicted values
y_predict = reg.predict(x_test)
print(y_predict)
```

```python
#actual values
print(y_test)
```

```python
#check model performance and the accuracy using mean_square method
print(np.mean((y_test-y_predict)**2))
```

```python
print(round(reg.score(x_test, y_test)*100, 2))
```
